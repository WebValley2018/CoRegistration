```python
dwi_image_np = sitk.GetArrayFromImage(dwi_image)
moving_image_np = sitk.GetArrayFromImage(moving_image)

#scipy.misc.imsave('/Users/riccardobusetti/Desktop/img.jpg', dwi_image_np[10,:,:])

def normalize(arr):
    rng = arr.max()-arr.min()
    amin = arr.min()
    return (arr-amin)*255/rng

def compare_images(img1, img2):
    img1 = normalize(img1)
    img2 = normalize(img2)
    diff = img1 - img2
    m_norm = sum(abs(diff))
    z_norm = norm(diff.ravel(), 0)
    return (m_norm, z_norm)

def calculate_difference_of_all(image):
    sum = 0
    
    for i in range(0, (image.shape[0] - 1), 2):
        n_m, n_z = compare_images(image[i,:,:], image[i+1,:,:])
        #print("{:,}".format(n_m))
        sum = sum + n_m
    
    return sum

def calculate_difference(image):
    n_m, n_z = compare_images(image[image.shape[0]//2,:,:], image[(image.shape[0]//2)+1,:,:])
    return n_m

def calculate_max_difference(image):
    diff = 0
    
    for i in range(0, (image.shape[0] - 1), 2):
        n_m, n_z = compare_images(image[i,:,:], image[i+1,:,:])
        if n_m > diff:
            diff = n_m
        
    return diff   

def calculate_difference_with_cross_sum_mean(image):
    sum = 0
    
    for i in range(0, (image.shape[0] - 2), 2):
        n_m, n_z = compare_images(image[i,:,:], image[i+2,:,:])
        sum = sum + n_m
        
    return int(sum / image.shape[0])

def calculate_difference_with_cross_min_diff(image):
    min = 9999999999
    
    for i in range(0, (image.shape[0] - 2), 2):
        n_m, n_z = compare_images(image[i,:,:], image[i+2,:,:])
        if (n_m < min):
            min = n_m
            
    return min

def print_image(image, series_number):
    for i in range(0, series_number):
        plt.imshow(image[i,:,:])
        plt.show()
        
def get_dicom_as_numpy(dir):
    reader = sitk.ImageSeriesReader()
    reader.SetOutputPixelType(sitk.sitkFloat32)
    dicom_names = reader.GetGDCMSeriesFileNames(dir)
    reader.SetFileNames(dicom_names)
    return sitk.GetArrayFromImage(reader.Execute())
        
for i in range(0, 9):
    image_np = get_dicom_as_numpy(f"/Users/riccardobusetti/Desktop/tmp_processed/pa001/st000/se00{i}")
    print(f"Image {i} difference ", "{:,}".format(calculate_difference_with_cross_sum_mean(image_np))) 
    
image = get_dicom_as_numpy("/Users/riccardobusetti/Desktop/tmp_processed/pa001/st000/se008")
print_image(image, image.shape[0])
```
